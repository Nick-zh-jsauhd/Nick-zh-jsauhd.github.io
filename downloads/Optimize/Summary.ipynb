# 优化模型
### 优化模型的一般意义
优化模型是数学建模竞赛中出现频率最高的模型类型，据统计，进五年国赛题目中约有35%的问题需要用到优化思想

优化问题，一般是指用“最好”的方式，使用或分配有限的资源，即劳动力，原材料，机器，资金等，使得费用最小或利润最大

人们在解决实际问题时往往会提出若干方案，通过各方面的信息论证，从中提取最佳方案。而在优化问题中，我们关心如何从多个方案中科学合理地提取出最佳方案。它包含两个方面的内容：
- **建立数学模型** - 模型中的数学关系反应最优化问题中的所要达到的目标和各种约束条件。
- **求解** - 数学模型建立完成之后，选择合理的方式进行求解
---
### 优化模型的数学描述形式

将一个优化问题用数学式子来描述，即**求函数**

$$
u = f(x), \quad x = (x_1, x_2, x_3, \ldots, x_n)
$$

在约束条件

$$
h_i(x) = 0, \quad i = 1, 2, \ldots, m
$$

和

$$
g_i(x) \leq 0 \; (g_i(x) \geq 0), \quad i = 1, 2, \ldots, p
$$

下的**最大值或最小值**，其中：

- $ x $ → $ x = (x_1, x_2, \ldots, x_n)^T $ **决策变量**
- $ f(x) $ → **目标函数**
- $ x \in \Omega $ → **可行域**

$ \min(or \max)u = f(x), x \in \Omega $

s.t. $ h_i(x) = 0, \quad i = 1, 2, \ldots, m $

$ g_i(x) \leq 0 \; (g_i(x) \geq 0), \quad i = 1, 2, \ldots, p $ 

--- 
### 优化模型的三要素 -> 决策， 目标， 约束
- **决策变量** -该问题要求解的那些未知变量
- **目标函数** -通常是该问题要优化（最大或最小）的那个目标的数学表达式，他是决策变量的函数
- **约束条件** -由该问题对决策变量的限制条件给出

--- 
### 优化模型的分类
1. **根据是否存在约束条件** —— 有约束问题与无约束问题
2. **根据设计变量的性质** —— 静态问题与动态问题
3. **根据目标函数和约束条件表达式的性质** —— 线性规划，非线性规划，二次规划， 多目标规划等
